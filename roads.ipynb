```python
%%sql
-- Check the database works
SELECT * from Street_Names
LIMIT 10; 
```

```python
%%sql
-- What are the most common street suffixes?
SELECT StreetSuffix,count(StreetSuffix) as occurances from Street_Names 
where StreetSuffix is not null 
and StreetType is "PUBLIC STREET NAMES" 
GROUP by StreetSuffix 
order by occurances desc;
```

```python
%%sql
-- What are the most common street names?
SELECT StreetName,count(StreetName) as occurances from Street_Names 
where StreetName is not null 
and StreetType = "PUBLIC STREET NAMES" 
GROUP by StreetName
order by occurances desc
```

```python
%%sql
--What/Where are the SHERMAN Streets?
SELECT OfficialStreetName, ThomasBrothersMap
FROM Street_Names
WHERE StreetName = "SHERMAN"
AND StreetType = "PUBLIC STREET NAMES";
```

```python
%%sql
-- How about PARK?
SELECT OfficialStreetName, ThomasBrothersMap
FROM Street_Names
WHERE StreetName = "PARK"
AND StreetType = "PUBLIC STREET NAMES";
```

```python

%%sql
-- What are the most common street names, including private street names, etc. ?
SELECT StreetSuffix, count(StreetSuffix) as occurances from Street_Names 
where StreetSuffix is not null 
GROUP by StreetSuffix 
order by occurances desc;
```

```python
%%sql
--Where are the 27TH Streets?
SELECT OfficialStreetName, ThomasBrothersMap
FROM Street_Names
WHERE StreetName = "27TH"
AND StreetType = "PUBLIC STREET NAMES";
```
